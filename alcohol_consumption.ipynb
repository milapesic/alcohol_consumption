```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df.dropna(subset=['Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)', 'GDP per capita, PPP (constant 2017 international $)','Population (historical estimates)'],inplace=True)
df
```

```python
pivot=df.pivot(index='Year', columns='Entity', values='Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)')
pivot
```

```python
pivot_table = pd.pivot_table(df, values='Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)', index=['Entity'],
                    aggfunc=np.mean)
pivot_table
```

```python
pivot_table_sorted=pivot_table.sort_values(by='Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)',ascending=False)
```

```python
top10=pivot_table_sorted.head(10)
top10
```

```python
top10
```

```python
top10.plot(kind='bar',xlabel='Country',ylabel='Average alcohol consumption in liters of pure alcohol per capita (2000-2018)',title='Top 10 Countries with the Highest Alcohol Consumption (2000-2018)',figsize=(12,6),color='#fc080c')

plt.show()
```

```python
pivot_selected=pivot[['Lithuania','Moldova','Czechia','Belarus','Russia','Romania','Uganda','Luxembourg','Ireland','Germany']]
```

```python
pivot_selected.plot(kind='line',xlabel='Year',ylabel='Average alcohol consumption (in liters of pure alcohol per capita)',title='Alcohol Consumption in top 10 Countries with the Highest Alcohol Consumption',figsize=(12,6),)
plt.legend(prop={"size":8})
plt.xticks(pivot_selected.index)
plt.show()
```

```python
last10=pivot_table_sorted.tail(10)
last10
```

```python
df['Total consumption (liters of pure alcohol)']=df.apply(lambda row:(row['Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)']*row['Population (historical estimates)']),axis=1)
```

```python
df['Entity and year']=df.apply(lambda row:(row['Entity']+' '+str(row['Year'])),axis=1)
```

```python
df
```

```python
df=df[df['Entity and year'].str.contains('World')==False]
df
```

```python
table=df.groupby('Year',as_index=False)['Total consumption (liters of pure alcohol)'].sum()
table
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x = list(table.Year)
y=list(table['Total consumption (liters of pure alcohol)'])
ax.bar(x,y,color='#7eb54e',width=1.5)
plt.ylabel('Alcohol consumption (liters of pure alcohol)', fontsize=12)
plt.xlabel('Year', fontsize=12)
plt.xticks(fontsize=12,rotation='vertical')
plt.yticks(y,fontsize=8)
plt.xticks(list(table['Year']))
plt.title('Total Alcohol Consumption in the World (2000-2018)')
fig.set_size_inches(8,6)
plt.show()
```

```python
correlation=df['Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)'].corr(df['GDP per capita, PPP (constant 2017 international $)'])
print('Pearson correlation coefficient is',correlation,'which means:')
if correlation==1:
    print('Perfect positive linear relationship')
elif correlation<1 and correlation>0:
    print('Positive correlation')
elif correlation==0:
    print('No linear relationship')
elif correlation<0 and correlation>-1:
    print('Negative correlation')
else:
    print('Perfect negative linear relationship')
```
